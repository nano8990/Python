# 지도학습(Regression)

```python
import numpy as np
import pandas as pd

# 모델 라이브러리 선언
from sklearn import datasets, tree

# 모델 정확도 라이브러리 선언
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

# CSV 파일을 읽어 DataFrame 변수에 저장하기
featuresData = pd.read_csv("../Python_ST_EX/dataset/feature_regression_example.csv")
featuresData.head()
```

## 1. 타입 통합 / 특성 숫자컬럼 추가

### [실습] 데이터 타입을 모델정의서와 같이 설정하며 홀리데이 및 프로모션 여부 컬럼에 대해서 Y→1, N→0 컬럼을 추가로 생성하세요

```python
featuresData["HOLIDAY_YN"] = np.where(featuresData["HOLIDAY"] == "Y", 1, \
                                       np.where(featuresData["HOLIDAY"] == "N", 0, \
                                                "ERROR")).astype(int)
featuresData["PROMOTION_YN"] = np.where(featuresData["PROMOTION"] == "Y", 1, \
                                   np.where(featuresData["PROMOTION"] == "N", 0, \
                                            "ERROR")).astype(int)
featuresData.head()
```

### [실습] 데이터 범위를 YEARWEEK 기준 201501 ~ 201652 구간만 조회하여 FeaturesData 변수에 저장하세요.

```python
featuresData2 = featuresData[ (featuresData["YEARWEEK"] >= 201501) & (featuresData["YEARWEEK"] <= 201652) ]
featuresData2.head()
```

## 2. 특성선정 / 데이터 분리

```python
# 상관관계 확인
corrDf = featuresData2.corr()
corrDf
```

```python
# 인덱스컬럼 초기화
corrResult = corrDf.reset_index()
corrResult
```

```python
# 상관관계 계수 비율 선정
featuresStd = 0.5

# Feature 선택
features = corrResult[ (corrResult["QTY"] != 1) & (abs(corrResult["QTY"]) > featuresStd) ]["index"].tolist()
features
```

```python
label = ["QTY"]
label
```

```python
# 데이터 분리
yearweekStd = 201630

trainingData_features = featuresData2[featuresData2.YEARWEEK <= yearweekStd][features].reset_index(drop = True)
trainingData_label = featuresData2[featuresData2.YEARWEEK <= yearweekStd][label].reset_index(drop = True)
testData_features = featuresData2[featuresData2.YEARWEEK > yearweekStd][features].reset_index(drop = True)
testData = featuresData2[featuresData2.YEARWEEK > yearweekStd].reset_index(drop = True)
```

## 3. 모델 적용

```python
# Define Model
model_method_classifier = tree.DecisionTreeClassifier()
model_method_regressor = tree.DecisionTreeRegressor()
model_method_linear = LinearRegression()

# Learning (Feature & Label)
model_classifier = model_method_classifier.fit(trainingData_features, trainingData_label)
model_regressor = model_method_regressor.fit(trainingData_features, trainingData_label)
model_linear = model_method_linear.fit(trainingData_features, trainingData_label)
```

## 4. 예측

```python
predict_classifier = model_classifier.predict(testData_features)
predict_regressor = model_regressor.predict(testData_features)
predict_linear = model_linear.predict(testData_features)
predictDataClassifier = pd.DataFrame(predict_classifier, columns = ["PREDICT_CLASSIFIER"])
predictDataRegressor = pd.DataFrame(predict_regressor, columns = ["PREDICT_REGRESSOR"])
predictDataLinear = pd.DataFrame(predict_linear, columns = ["PREDICT_LINEAR"])
```

## 5. 데이터 정리

```python
verificationData = pd.concat([testData, predictDataClassifier, predictDataRegressor, predictDataLinear], axis=1)
verificationData.head()
```

## 6. 결과 검증

```python
import math
### MAE
dataMAE_classifier = mean_absolute_error(verificationData['QTY'], verificationData['PREDICT_CLASSIFIER'])
dataMAE_regressor = mean_absolute_error(verificationData['QTY'], verificationData['PREDICT_REGRESSOR'])
dataMAE_linear = mean_absolute_error(verificationData['QTY'], verificationData['PREDICT_LINEAR'])
### RMSE (MSE)
dataRMSE_classifier = math.sqrt(mean_squared_error(verificationData['QTY'], verificationData['PREDICT_CLASSIFIER']))
dataRMSE_regressor = math.sqrt(mean_squared_error(verificationData['QTY'], verificationData['PREDICT_REGRESSOR']))
dataRMSE_linear = math.sqrt(mean_squared_error(verificationData['QTY'], verificationData['PREDICT_LINEAR']))
### R2
dataR2_classifier = r2_score(verificationData['QTY'], verificationData['PREDICT_CLASSIFIER'])
dataR2_regressor = r2_score(verificationData['QTY'], verificationData['PREDICT_REGRESSOR'])
dataR2_linear = r2_score(verificationData['QTY'], verificationData['PREDICT_LINEAR'])
```

```python
print("%s\t%s\t%s\n%s\t%s\t%s\n%s\t%s\t%s" %(dataMAE_classifier, dataMAE_regressor, dataMAE_linear, dataRMSE_classifier, dataRMSE_regressor, dataRMSE_linear, dataR2_classifier, dataR2_regressor, dataR2_linear))
```

## DECISION TREE

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

```python
# Define Model
model_method = tree.DecisionTreeRegressor()

# Learning (Feature & Label)
model = model_method.fit(trainingData_features, trainingData_label)
```

```python
dot_data = StringIO()
export_graphviz(model, out_file = dot_data, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
