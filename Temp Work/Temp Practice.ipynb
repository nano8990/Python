```python
def dataManipulation(dataFrame, movingAverage, groupList = None, aggColumn = None, aggFunc = 'mean', sortList = None, ascending = True):
    import math
    
    if (sortList != None):
        sortedDataFrame = dataFrame.sort_values(sortList, ascending = ascending, inplace = False)
    else:
        sortedDataFrame = dataFrame
        
    if (groupList != None):
        if (aggColumn != None):
            groupedDataFrame = sortedDataFrame.groupby(groupList).agg(aggFunc)[aggColumn].reset_index()
        else:
            groupedDataFrame = sortedDataFrame.groupby(groupList).agg(aggFunc).reset_index()
    else:
        groupedDataFrame = sortedDataFrame

    def movingAverageCalculation(sameGroup):
        updownRange = math.floor(movingAverage / 2)
        movingAverageList = []
        for i in range(sameGroup[0], sameGroup[len(sameGroup)-1] + 1):
            perMovingAverage = [groupedDataFrame[aggColumn][i]]
            lengthCount = 1
            for j in range(1, updownRange + 1):
                if (i - j > sameGroup[0]):
                    perMovingAverage += [groupedDataFrame[aggColumn][i - j]]
                    lengthCount += 1
                if (i + j < sameGroup[len(sameGroup) - 1]):
                    perMovingAverage += [groupedDataFrame[aggColumn][i + j]]
                    lengthCount += 1
            movingAverageList += [sum(perMovingAverage) / lengthCount]
        return movingAverageList

    movingAverageTotalList = []
    groupingBoolean = True
    for i in range(0, len(groupedDataFrame) - 1):
        tempList1 = []
        tempList2 = []
        for j in range(0, len(groupList)):
            tempList1 += groupedDataFrame[groupList[j]][i]
            tempList2 += groupedDataFrame[groupList[j]][i + 1]
        if (groupingBoolean == True):
            sameGroup = [i]
        if (tempList1 == tempList2 and i == len(groupedDataFrame) - 2):
            movingAverageTotalList += movingAverageCalculation(sameGroup)
            groupingBoolean = True
        elif (tempList1 == tempList2):
            sameGroup += [i + 1]
            groupingBoolean = False
        else:
            if (i == len(groupedDataFrame) - 2):
                movingAverageTotalList += movingAverageCalculation(sameGroup)
                sameGroup = [len(groupedDataFrame) - 1]
                movingAverageTotalList += movingAverageCalculation(sameGroup)
            else:
                movingAverageTotalList += movingAverageCalculation(sameGroup)
            groupingBoolean = False

    groupedDataFrame['MA'] = movingAverageTotalList
    result = groupedDataFrame
    return result
```
