# 1. 데이터 시각화하기(matplotlib)

```python
# 필요 라이브러리 정의
from matplotlib import pyplot as plt
# 커맨드뷰에서 차트 시연
%matplotlib inline
```

```python
x = list(range(1, 11, 1))
y = []
for i in x:
    y.append(i ** 2)
print(x, y)
```

```python
# linewidth : 선 굵기
# markersize : 점 크기
# shift + tab으로 가이드를 받아서 인자를 추가하면 됨
# 색, 점(o), 선(-)
plt.plot(x, y, 'g*-', markersize = 40, linewidth = 5)
# 차트 꾸미기
plt.xlabel('index') # x축 라벨
plt.ylabel('value') # y축 라벨
plt.title('Simple plot') # 차트 제목
# 차트 꾸미기, 특정구간 확대
# plt.xlim(3,7)
# plt.ylim(10,40)
```

```python
"""
loc : int or string or pair of floats, default: :rc:`legend.loc` ('best' for axes, 'upper right' for figures)
The location of the legend. Possible codes are:
===============   =============
Location String   Location Code
===============   =============
'best'            0
'upper right'     1
'upper left'      2
'lower left'      3
'lower right'     4
'right'           5
'center left'     6
'center right'    7
'lower center'    8
'upper center'    9
'center'          10
===============   =============
"""
# 레전드 설정
x=list(range(1, 11, 1))
y = [i**2 for i in x] # 파이썬에서 사용 가능한 for문의 새로운 형태
y2 = [i**2.2 for i in x]
plt.plot(x, y, 'bo--',label = 'first')
plt.plot(x, y2, 'ro--', label = 'second')
plt.legend(loc = 0)
# 어노테이션
plt.annotate('point',xy=(2,8), xytext=(3,40),arrowprops=dict(color="blue", width=2))
```

```python
# 종합
plt.plot(x, y, 'bo-', linewidth = 1.5)
plt.grid(True) #그리드 표기
plt.xlabel('index') # x축 라벨
plt.ylabel('value') # y축 라벨
plt.title('Simple plot') # 차트 제목
```

# 2. 데이터 시각화 고급

## 1) 서브플롯 그리기

```python
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

# 서브플롯
fig = plt.figure(figsize=(10,5)) # 도화지 크기 설정
ax1 = fig.add_subplot(211) # 행열의 개념 / 0행 0열의 0번째
plt.plot(x,y, 'b',lw=1.5, label = '1st')
plt.plot(x,y, 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')
ax2 = plt.subplot(212)
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
plt.plot(x,y2, 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

## 2) 이중축 활용하기

```python
fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(1,1,1)
plt.plot(x, y, 'bo--', linewidth = 1.5, label = '1st')
plt.grid(True)
plt.legend(loc = 2)
plt.ylabel('value')
plt.title('subplot')
ax2 = ax1.twinx()
plt.plot(x, y2, 'go-', linewidth = 1.5, label = '2nd')
plt.legend(loc = 1)
plt.xlabel('index')
plt.ylabel('value')
```

## 1) CountPlot 구현

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
sns.countplot(x = "day", data = tips)
sns.load_dataset('tips').head()
```

## 2) Distplot 구현

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['tip'])
```

```python
sns.distplot(tips['size'])
```

## 3) Jointplot 구현

```python
# 2차원 실수형 데이터 시각화
sns.jointplot(x= 'total_bill', y = 'tip', data = tips, kind = 'kde')
# 데이터가 숫자형 데이터가 2차원으로 있는 경우 활용 가능
# kind = 'kde' 추가 시 커널 밀도함수 시연
```

## 6) boxplot 차트 구현

```python
# outlier
# 3사분위수 + (IQR * 1.5)
# 3사분위수 (75%)
# 중앙값
# 1사분위수 (25%)
# 1사분위수 - (IQR * 1.5)
# * IQR = 3사분위수 - 1사분위수
sns.boxplot(x='day', y='total_bill',data=tips)
```
