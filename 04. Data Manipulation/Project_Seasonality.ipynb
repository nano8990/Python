### [지역, 상품, 연도 별 집계] sortedData 에서 지역, 상품 단위 판매량(QTY_NEW) 의 평균 및 표준편차 연산 후 groupData 변수에 담으세요.
### 이후 컬럼명을 QTY_MEAN, QTY_STD로 변경하세요

```python
groupKey = ['REGIONID', 'PRODUCT', 'YEAR']
groupData = sortedData.groupby(groupKey).agg(['mean','std'])['QTY_NEW']
groupData = groupData.reset_index()
groupData.head()
```

```python
newColumnNames = []
for i in range(0, len(groupData.columns)):
    columnNames = groupData.columns[i]
    if(columnNames == "mean"):
        newColumnNames += ['QTY_MEAN']
    elif(columnNames == "std"):
        newColumnNames += ['QTY_STD']
    else:
        newColumnNames += [columnNames]
groupData.columns = newColumnNames
groupData.head()
```
