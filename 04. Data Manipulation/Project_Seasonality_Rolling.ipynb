```python
def project_seasonality_index(dataFrame, groupKey, aggColumn, movingAverage = [15, 9, 9], sortKey = None, ascending = True):   
    import math
    import numpy as np

    import timeit
    start = timeit.default_timer()
    # Your code start    
    
    if (sortKey != None):
        sortedDataFrame = dataFrame.sort_values(sortKey, ascending = ascending, inplace = False)
    else:
        sortedDataFrame = dataFrame
        
    groupedDataFrame = sortedDataFrame.reset_index(drop = True)
    
    rowLength = len(groupedDataFrame)
    
    stop = timeit.default_timer()
    print('Time: ' , stop - start, 'DataSort')
    
    start = timeit.default_timer()
    # Your code start
    
    groupListForMatch = [""] * rowLength
    for i in groupKey:
        groupListForMatch += groupedDataFrame[i].astype(str)
    sameGroup = []
    for i in range(0, rowLength):
        if (i == 0):
            matchGroup1 = groupListForMatch[i]
            sameGroup = [i]
        else:
            matchGroup2 = groupListForMatch[i]
        if (i > 0):
            if (matchGroup1 != matchGroup2):
                matchGroup1 = matchGroup2
                sameGroup += [i]
    
    sameGroupLength = len(sameGroup)
    
    updownRange = math.floor(movingAverage[0] / 2)

    stop = timeit.default_timer()
    print('Time: ' , stop - start, 'Grouping')

    start = timeit.default_timer()
    
    aggColumnList = list(groupedDataFrame[aggColumn])
    
    movingAverageList = []
    for i in range(0, sameGroupLength):
        if (i < sameGroupLength - 1):
            perMovingAverage = pd.DataFrame({"MA_TEMP": aggColumnList[sameGroup[i]:sameGroup[i+1]]})
            perMovingAverageRolling = perMovingAverage.MA_TEMP.rolling(window = 15, center = True, min_periods = 1).mean()
        else:
            perMovingAverage = pd.DataFrame({"MA_TEMP": aggColumnList[sameGroup[i]:rowLength]})
            perMovingAverageRolling = perMovingAverage.MA_TEMP.rolling(window = 15, center = True, min_periods = 1).mean()
        movingAverageList += list(perMovingAverageRolling)
    
    columnName = 'MA'
    columnNameMatch = columnName
    tempCount = 1
    while(list(groupedDataFrame.columns).count(columnNameMatch) > 0):
        tempCount += 1
        columnNameMatch = columnName + str(tempCount)
    columnName = columnNameMatch
    groupedDataFrame[columnName] = movingAverageList

    stop = timeit.default_timer()
    print('Time: ' , stop - start, 'MovingAverage')
    
    start = timeit.default_timer()
    
    updownRange = math.floor(movingAverage[1] / 2)

    movingStdList = []
    for i in range(0, sameGroupLength):
        if (i < sameGroupLength - 1):
            perMovingStd = pd.DataFrame({"MSTD_TEMP": movingAverageList[sameGroup[i]:sameGroup[i+1]]})
            perMovingStdRolling = perMovingStd.MSTD_TEMP.rolling(window = 9, center = True, min_periods = 1).std(ddof = 1)
        else:
            perMovingStd = pd.DataFrame({"MSTD_TEMP": movingAverageList[sameGroup[i]:rowLength]})
            perMovingStdRolling = perMovingStd.MSTD_TEMP.rolling(window = 9, center = True, min_periods = 1).std(ddof = 1)
        movingStdList += list(perMovingStdRolling)
    
    columnName = 'MSTD'
    columnNameMatch = columnName
    tempCount = 1
    while(list(groupedDataFrame.columns).count(columnNameMatch) > 0):
        tempCount += 1
        columnNameMatch = columnName + str(tempCount)
    columnName = columnNameMatch
    groupedDataFrame[columnName] = movingStdList

    stop = timeit.default_timer()
    print('Time: ' , stop - start, 'MovingStandardDeviation')

    start = timeit.default_timer()
    
    refinedDataList = []
    for i in range(0, rowLength):
        if (aggColumnList[i] >= movingAverageList[i] + movingStdList[i]):
            refinedDataList += [movingAverageList[i] + movingStdList[i]]
        elif (aggColumnList[i] <= movingAverageList[i] - movingStdList[i]):
            refinedDataList += [movingAverageList[i] - movingStdList[i]]
        else:
            refinedDataList += [aggColumnList[i]]

    columnName = 'REFINED_' + aggColumn
    columnNameMatch = columnName
    tempCount = 1
    while(list(groupedDataFrame.columns).count(columnNameMatch) > 0):
        tempCount += 1
        columnNameMatch = columnName + str(tempCount)
    columnName = columnNameMatch
    groupedDataFrame[columnName] = refinedDataList
    
    stop = timeit.default_timer()
    print('Time: ' , stop - start, 'DataRefine')
    
    start = timeit.default_timer()
    
    updownRange = math.floor(movingAverage[2] / 2)
 
    trendlineList = []
    for i in range(0, sameGroupLength):
        if (i < sameGroupLength - 1):
            perTrendline = pd.DataFrame({"TRENDLINE_TEMP": refinedDataList[sameGroup[i]:sameGroup[i+1]]})
            perTrendlineRolling = perTrendline.TRENDLINE_TEMP.rolling(window = 9, center = True, min_periods = 1).mean()
        else:
            perTrendline = pd.DataFrame({"TRENDLINE_TEMP": refinedDataList[sameGroup[i]:rowLength]})
            perTrendlineRolling = perTrendline.TRENDLINE_TEMP.rolling(window = 9, center = True, min_periods = 1).mean()
        trendlineList += list(perTrendlineRolling)
    
    columnName = 'TRENDLINE'
    columnNameMatch = columnName
    tempCount = 1
    while(list(groupedDataFrame.columns).count(columnNameMatch) > 0):
        tempCount += 1
        columnNameMatch = columnName + str(tempCount)
    columnName = columnNameMatch
    groupedDataFrame[columnName] = trendlineList

    stop = timeit.default_timer()
    print('Time: ' , stop - start, 'Trendline')
    
    start = timeit.default_timer()
    
    seasonalityIndex1List = []
    for i in range(0, rowLength):
        if (aggColumnList[i] > 0):
            seasonalityIndex1List += [aggColumnList[i] / trendlineList[i]]
        else:
            seasonalityIndex1List += [1]

    columnName = 'SEASONALITY_INDEX1'
    columnNameMatch = columnName
    tempCount = 1
    while(list(groupedDataFrame.columns).count(columnNameMatch) > 0):
        tempCount += 1
        columnNameMatch = columnName + str(tempCount)
    columnName = columnNameMatch
    groupedDataFrame[columnName] = seasonalityIndex1List
    
    seasonalityIndex2List = []
    for i in range(0, rowLength):
        if (refinedDataList[i] > 0):
            seasonalityIndex2List += [refinedDataList[i] / trendlineList[i]]
        else:
            seasonalityIndex2List += [1]

    columnName = 'SEASONALITY_INDEX2'
    columnNameMatch = columnName
    tempCount = 1
    while(list(groupedDataFrame.columns).count(columnNameMatch) > 0):
        tempCount += 1
        columnNameMatch = columnName + str(tempCount)
    columnName = columnNameMatch
    groupedDataFrame[columnName] = seasonalityIndex2List
    
    stop = timeit.default_timer()
    print('Time: ' , stop - start, 'SeasonalityIndex')
    
    return groupedDataFrame
```

```python
def project_seasonality_index(dataFrame, groupKey, aggColumn, movingSection = [15, 9, 9], sortKey = None, ascending = True):   
    # 필요함수 호출
    import math
    import numpy as np
    
    # 변수명 중복을 처리하는 함수 정의
    def columnName_function(dataFrame, columnName):
        columnNameMatch = columnName
        tempCount = 1
        while(list(dataFrame.columns).count(columnNameMatch) > 0):
            tempCount += 1
            columnNameMatch = columnName + "_" + str(tempCount)
        columnName = columnNameMatch
        return columnName
    
    # 그룹별 이동평균, 변동률, 추세선을 구하는 함수 정의
    def rolling_function(dataFrame, groupKey, rollingColumnName, window, aggFunc):
        groupDataSet = dataFrame.groupby(groupKey)
        groupDataSetList = list(groupDataSet.groups)
        rollingList = []
        for i in range(0, len(groupDataSetList)):
            rollingValue = groupDataSet.get_group(groupDataSetList[i])
            if (aggFunc == 0):
                rollingList += list(rollingValue[rollingColumnName].rolling(window = window, center = True, min_periods = 0).mean())
            else:
                rollingList += list(rollingValue[rollingColumnName].rolling(window = window, center = True, min_periods = 0).std(ddof = 1))
        return rollingList
    
    # 데이터 정렬
    if (sortKey != None):
        sortedDataFrame = dataFrame.sort_values(sortKey, ascending = ascending, inplace = False)
    else:
        sortedDataFrame = dataFrame
    
    # 연산 필요 변수 초기 설정
    groupedDataFrame = sortedDataFrame.reset_index(drop = True)
    aggColumnList = list(groupedDataFrame[aggColumn])
    rowLength = len(groupedDataFrame)
    
    # 이동평균 (15구간 평균)
    movingAverageList = rolling_function(dataFrame = groupedDataFrame, groupKey = groupKey, rollingColumnName = aggColumn, window = movingSection[0], aggFunc = 0)
    columnNameMovingAverage = columnName_function(groupedDataFrame, "MA")
    groupedDataFrame[columnNameMovingAverage] = movingAverageList

    # 변동률 (9구간 표준편차)
    movingStdList = rolling_function(dataFrame = groupedDataFrame, groupKey = groupKey, rollingColumnName = columnNameMovingAverage, window = movingSection[1], aggFunc = 1)
    columnNameMovingStd = columnName_function(groupedDataFrame, "MSTD")
    groupedDataFrame[columnNameMovingStd] = movingStdList
    
    # 데이터 정제 (데이터의 상한선, 하한선 설정)
    refinedDataList = []
    for i in range(0, rowLength):
        upperBound = movingAverageList[i] + movingStdList[i]
        lowerBound = movingAverageList[i] - movingStdList[i]
        if (aggColumnList[i] >= upperBound):
            refinedDataList += [upperBound]
        elif (aggColumnList[i] <= lowerBound):
            refinedDataList += [lowerBound]
        else:
            refinedDataList += [aggColumnList[i]]
    columnNameRefinedValue = columnName_function(groupedDataFrame, "REFINED_" + aggColumn)
    groupedDataFrame[columnNameRefinedValue] = refinedDataList
    
    # 추세선 (정제된 데이터의 9구간 이동평균)
    trendlineList = rolling_function(dataFrame = groupedDataFrame, groupKey = groupKey, rollingColumnName = columnNameRefinedValue, window = movingSection[2], aggFunc = 0)
    columnNameTrendline = columnName_function(groupedDataFrame, "TRENDLINE")
    groupedDataFrame[columnNameTrendline] = trendlineList
    
    # 계절성지수1 (원래 데이터 / 추세선 ; 안정된 시장)
    seasonalityIndex1List = []
    for i in range(0, rowLength):
        if (aggColumnList[i] > 0 and trendlineList[i] > 0):
            seasonalityIndex1List += [aggColumnList[i] / trendlineList[i]]
        else:
            seasonalityIndex1List += [1]
    columnNameSeasonalityIndex1 = columnName_function(groupedDataFrame, "SEASONALITY_INDEX(STABLE_MARKET)")
    groupedDataFrame[columnNameSeasonalityIndex1] = seasonalityIndex1List
    
    # 계절성지수2 (정제된 데이터 / 추세선 ; 불안정 시장)
    seasonalityIndex2List = []
    for i in range(0, rowLength):
        if (refinedDataList[i] > 0 and trendlineList[i] > 0):
            seasonalityIndex2List += [refinedDataList[i] / trendlineList[i]]
        else:
            seasonalityIndex2List += [1]
    columnNameSeasonalityIndex2 = columnName_function(groupedDataFrame, "SEASONALITY_INDEX(UNSTABLE_MARKET)")
    groupedDataFrame[columnNameSeasonalityIndex2] = seasonalityIndex2List

    return groupedDataFrame
```
