```python
while(True):
    host='192.168.110.116'
    for i in range(8023, 8044):
        mysock=socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        try:
            print("try : " + str(i))
            addr=(host,i)
            mysock.connect(addr)
        except:
            print("except : " + str(i))
            continue

        try:
            msg=b"DongJoon JJang JJang Man DongJoon JJang JJang Man DongJoon JJang JJang Man DongJoon JJang JJang Man\n"
            try:
                for i in range(0, 100):
                    mysock.sendall(msg)
            except:
                continue
        except socket.errno as e:
            print("Socket error ", e)
        finally:
            mysock.close()
```
