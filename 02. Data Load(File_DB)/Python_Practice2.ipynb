#### 2. 판매량 List 불러오기

```python
import timeit
start = timeit.default_timer()
# Your code start
import pandas as pd
qtyList = pd.read_csv("../Python_ST_EX/dataset/kopo_channel_seasonality_new.csv")["QTY"].tolist()
cntMax = qtyList.count(max(qtyList))
cntMin = qtyList.count(min(qtyList))
for i in range(0,cntMax):
    del qtyList[qtyList.index(max(qtyList))]
for i in range(0,cntMin):
    del qtyList[qtyList.index(min(qtyList))]
average = sum(qtyList) / len(qtyList)
print(average)
# Your code end
stop = timeit.default_timer()
print('Time: ' , stop - start)
```

# 3. 데이터 불러오기 from 데이터베이스 (Oracle)

```python
# DB연동 패키지 설치필요
# pip install cx_Oracle
# Oracle client 설치 필요 Toad 설치 후 tnsname 설정 후 재시작 필요

import pandas as pd
from sqlalchemy import create_engine

# DB 커넥션 열기
engine_Oracle = create_engine('oracle+cx_oracle://kopo2019:kopo2019@192.168.110.112:1521/orcl') # ID : PW @ IP : PORT / SERVICE NAME

# DB 테이블을 읽어 Data Frame 변수에 저장하기
selloutData = pd.read_sql_query('SELECT * FROM kopo_product_volume', engine_Oracle)

# 컬럼헤더 재정의
selloutData.columns = ['SalesID','ProductGroup','YearWeek','Volume']

# 데이터 VIEW
selloutData.head()

# Oracle 데이터 저장 속도 향상방법
from sqlalchemy import types

# 데이터 저장
resultname = 'oracleresult_KimDongJoon'

to_varchar = {c:types.VARCHAR(selloutData[c].str.len().max()) \
              for c in selloutData.columns[selloutData.dtypes == 'object'].tolist()}

selloutData.to_sql(resultname, engine_Oracle, if_exists = 'replace', index = False, dtype = to_varchar)

# 데이터 저장 Tip
# 오라클은 저장시 selloutData.dtypes를 확인하면 object 속성이 추후 clob 형태로 저장된다.
# 따라서 문자열형태로 변환을 한 후 저장시 더 빠른 속도를 낼 수 있다.
```

```python
engine_Oracle = create_engine('oracle+cx_oracle://DJ2019:DJ2019@192.168.110.3:1522/XE') # ID : PW @ IP : PORT / SERVICE NAME

selloutData = pd.read_sql_query('SELECT * FROM KOPO_PRODUCT_VOLUME', engine_Oracle)

selloutData.columns = ['SalesID','ProductGroup','YearWeek','Volume']

resultname = 'oracleresult'
selloutData.to_sql(resultname, engine_Oracle, if_exists = 'replace', index = True)
selloutData.head()
```

```python
# Pandas 패키지 불러오기
import pandas as pd
import pymysql
from sqlalchemy import create_engine

# DB 커넥션 열기
engine_mysql = create_engine('mysql+pymysql://kopo:kopo@13.209.112.251:3306/kopo') # ID : PW @ IP : PORT / DB
engine_postgresql = create_engine('postgresql://kopo:kopo@192.168.110.111:5432/kopo')

# DB 테이블을 읽어 Data Frame 변수에 저장하기
selloutData = pd.read_sql_query('SELECT * FROM kopo_product_volume', engine_mysql)

# 컬럼헤더 재정의
selloutData.columns = ['AA','BB','CC','DD']

# 데이터 저장
resultname = 'kopo_product_volume_김동준'
selloutData.to_sql(resultname, engine_postgresql, if_exists = 'replace', index = False)
selloutData.head()

# 컬럼헤더 재정의2
selloutData.columns = ['aa','bb','cc','dd']

# 데이터 저장2
resultname = 'kopo_product_volume_김동준2'
selloutData.to_sql(resultname, engine_postgresql, if_exists = 'replace', index = False)
selloutData.head()
```
