```python
# PANDAS 패키지 불러오기
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# CSV 파일을 읽어 Data Frame 변수에 저장하기
studentData = pd.read_csv("../Python_ST_EX/dataset/student_middle_result.csv")
```

```python
name = studentData[ ['ST_NO']]
feature = studentData[ ['SW','DB','AND']]
```

```python
# create model and prediction
model = KMeans(n_clusters=2,algorithm='auto')
model.fit(feature)
predict = pd.DataFrame(model.predict(feature))
predict.columns=['PREDICT']
```

```python
# concatenate labels to df as a new column
r = pd.concat([name,feature,predict],axis=1)

# scatter plot
plt.scatter(r['DB'], r['AND'],c=r['PREDICT'],alpha=0.5)
# plt.scatter(center_x,center_y,s=50,marker='D',c='r')
plt.show()
```

```python
# 파이썬 사칙연산
integer_value = 123
float_value = 12.456

# 1. 더하기 연산
p_sum = integer_value+float_value
print(p_sum)

# 2. 곱하기 연산
p_multiple = integer_value*float_value
print(p_multiple)

# 3. 나누기 연산
p_division = integer_value/float_value
print(p_division)

# 4. 제곱연산
integer_value2 = int(float_value)
p_exponential = integer_value2**2
print(p_exponential)

# 5. 나머지 연산
p_mod = integer_value%int(float_value)
print(p_mod)
```

### 문자열 실습

```python
# 이것은 하드코딩! 실제로 쓰는 것은 지양해야 함
test = "SEC 20180212 250"
stock_name = test[:3]
date = test[4:4+8]
value = test[4+8+1:4+8+1+3]
print(stock_name); print(date); print(value)
```

```python
test1 = "SEC 20180212 250".split(" ")
stock_name = test1[0]
date = test1[1]
value = test1[2]
print(stock_name); print(date); print(value)
```

```python
# split을 이용한 문제풀이
test2 = "Sec 2018W28 250".upper().replace("W", "").split(" ")
stock_name = test2[0]
date = test2[1]
value = test2[2]
print(stock_name); print(date); print(value)
```

```python
# index를 이용한 문제풀이
test3 = "Sec 2018W28 250".upper().replace("W","")
del1 = test3.index(" ")
del2 = test3.index(" ", del1+1)
stock_name = test3[:del1]
date = test3[del1+1 : del2]
value = test3[del2+1:]
print(stock_name); print(date); print(value)
```

```python
# 교수님의 문제풀이
test2 = "Sec 2018W28 250"
UpperTest = test2.upper()
StdTest = UpperTest.replace("W","")
splitData = StdTest.split(" ")
print(test2); print(UpperTest); print(StdTest); print(splitData)
```

### 리스트 실습

```python
# 기본 함수로 풀기
test = [120, 150, 300, 500, 1000, 100, 100, 100, 100, 100, 100, 100, 100, 100, 2000, 2000, 2000, 2000, 2000, 2000, 2000, 2000, 2000, 2000]
cntMax = test.count(max(test))
cntMin = test.count(min(test))
for i in range(0,cntMax):
    del test[test.index(max(test))]
for i in range(0,cntMin):
    del test[test.index(min(test))]
average = sum(test) / len(test)
print(average)
```

```python
# Numpy 함수를 이용하여 풀기
import numpy as np
test = [120, 150, 300, 500, 1000, 100, 2000]
del test[test.index(max(test))]
del test[test.index(min(test))]
average = np.mean(test)
print(average)
```

```python
# 교수님의 풀이
# Value로는 이렇게 푼다
test = [120, 150, 300, 500, 1000, 100, 2000]
minValue = min(test)
maxValue = max(test)
test.remove(minValue)
test.remove(maxValue)
average = sum(test) / len(test)
print(average)

# 정렬로 풀기
test = [120, 150, 300, 500, 1000, 100, 2000]
test.sort()
del test[len(test)-1]
del test[0]
average = sum(test) / len(test)
print(average)
```
